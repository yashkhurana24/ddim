```python
grid1 = '/Users/yash/Desktop/ddim/ddim/model_outptus/diffusion/462_grid.jpeg'
grid2 = '/Users/yash/Desktop/ddim/ddim/model_outptus/gan/resized_grid.png'

# display both side by side
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].imshow(mpimg.imread(grid1))
ax[0].set_title('Diffusion')
ax[1].imshow(mpimg.imread(grid2))
ax[1].set_title('GAN')
plt.show()
```
